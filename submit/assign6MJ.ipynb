```python
import tensorflow as tf
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

im=tf.io.read_file('./datasets/ants/swiss-army-ant.jpg')
A=tf.io.decode_image(im)
K = tf.image.convert_image_dtype(A, tf.float32)
```

```python
K
```

```python
K1=tf.image.resize(K,[128, 128])
K1
tf.shape(K1)
```

```python
output = tf.image.random_crop(
   K1,  (64, 64, 3)
)
print(output)
plt.imshow(output)
```

```python
final=tf.image.resize_with_pad(output,128,128)
plt.imshow(final)
```
