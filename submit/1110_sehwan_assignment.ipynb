```python
# OpenCV, Matplotlib 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

#   
img = cv2.imread('./datasets/ants/swiss-army-ant.jpg')

#   
print(img.shape)

plt.imshow(img)
```

```python
img = cv2.resize(img, (128, 128))

print(img.shape)

plt.imshow(img)
```

```python
import tensorflow as tf 

img = tf.image.random_crop(img, size=[64, 64, 3])

print(img.shape)

plt.imshow(img)
```
