```python
# OpenCV, Matplotlib의 로드
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# 이미지 파일의 리드
img = cv2.imread('./datasets/ants/swiss-army-ant.jpg')

# 이미지 사이즈의 확인
print(img.shape)

plt.imshow(img)
```

```python
img = cv2.resize(img, (128, 128))

print(img.shape)

plt.imshow(img)
```

```python
import tensorflow as tf 

img = tf.image.random_crop(img, size=[64, 64, 3])

print(img.shape)

plt.imshow(img)
```
