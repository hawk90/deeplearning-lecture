```python
# Pandas의 로드
import pandas as pd

# 파일을 리드
Work_df = pd.read_csv('.\Friday-WorkingHours-Morning.pcap_ISCX.csv', sep=',')

# 선두에서 5행까지 표시
Work_df.head()
```

```python
#1. Flow ID열을 삭제하여라
Work_df.drop(['Flow ID'], axis=1,inplace = True)
Work_df.head()
```

```python
#2. Source IP, Destination IP는 inet_aton함수를 이용하여 숫자로 변환하여라
import socket
import struct

for i in Work_df[' Source IP']:
    SourceIP = struct.unpack("!I", socket.inet_aton(i))[0]
    print(SourceIP)
    
for j in Work_df[' Destination IP']:
    DestinationIP = struct.unpack("!I", socket.inet_aton(j))[0]
    print(DestinationIP)    
    
```
