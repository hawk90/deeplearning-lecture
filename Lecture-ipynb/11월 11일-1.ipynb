### 5-1．데이터의 이해
#### 데이터의 리드와 확인

```python
# OpenCV의 인스톨
!pip install opencv-python
```

```python
# OpenCV, Matplotlib의 로드
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# 이미지 파일의 리드
img = cv2.imread('./datasets/ants/swiss-army-ant.jpg')

# 이미지 사이즈의 확인
print(img.shape)

# 이미지 파일의 표시
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 이미지 파일의 표시
plt.imshow(img)
plt.show()
```

#### 픽셀값의 확인(컬러 이미지)

```python
# 이미지의 픽셀값을 확인
print(img)
```

```python
# 이미지의 배열사이즈를 확인
print(len(img)) # 세로 사이즈 y축
print(len(img[0])) # 가로 사이즈 x축
print(len(img[0][0])) # 컬러 채널
```

```python
# Pandas의 로드
import pandas as pd

# 픽셀배열을 Blue、Green、Red별로 분할
b, g, r = cv2.split(img)

# B배열을 데이터 프레임형식으로 변환
b_df = pd.DataFrame(b)

# 데이터 사이즈의 확인
print(b_df.shape)
# 데이터의 선두 5행을 표시
b_df.head()
```

```python
# 연습문제1
# G배열을 데이터 프레임으로 변환
g_df = pd.DataFrame(g)

# 데이터 사이즈의 확인
print(g_df.shape)
# 데이터의 선두 5행을 표시
g_df.head()
```

```python
# R配배열을 데이터 프레임으로 변환
r_df = pd.DataFrame(r)

# 데이터 사이즈의 확인
print(r_df.shape)
# 데이터의 선두 5행을 표시
r_df.head()
```

#### 픽셀값의 확인(그레이스케일 이미지)

```python
# 그레이스케일 이미지로 변환
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 이미지의 사이즈 확인
print(gray_img.shape)

# 이미지의 표시
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
# 연습문제2
# 이미지의 픽셀값을 확인
print(gray_img)
```

```python
# 연습문제3
# 이미지의 배열 사이즈를 확인
print(len(gray_img)) # 세로 사이즈 y축
print(len(gray_img[0])) # 가로 사이즈 x축
```

```python
# 배열을 데이터 프레임형식으로 변환
gr_df = pd.DataFrame(gray_img)

# 데이터 사이즈의 확인
print(gr_df.shape)
# 데이터의 선두 5행을 표시
gr_df.head()
```

#### 픽셀값의 확인(2치화 이미지)

```python
# 2치화이미지로 변환
ret, bin_img = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY)

# 이미지의 표시
plt.imshow(bin_img, cmap='gray')
plt.show()
```

```python
# 이미지의 픽셀값을 확인
print(bin_img)
```

```python
# 배열을 데이터 프레임형식으로 변환
bin_df = pd.DataFrame(bin_img)

# 데이터 사이즈의 확인
print(bin_df.shape)
# 데이터의 선두 5행을 표시
bin_df.head()
```
