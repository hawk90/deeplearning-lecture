```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_path = './datasets/ants/swiss-army-ant.jpg'
img_raw = tf.io.read_file(img_path)
img = tf.io.decode_image(img_raw)
img = tf.image.convert_image_dtype(img, tf.float32)
plt.imshow(img)
```

```python
resize_img = tf.image.resize(img, [128, 128])
plt.imshow(resize_img)
```

```python
crop_img = tf.image.random_crop(resize_img, size=[50, 50, 3])
plt.imshow(crop_img)
```

```python
tar_height = 0
try:
    pad_img = tf.image.resize_with_pad(crop_img, target_height=tar_height, target_width=256)
    plt.imshow(pad_img)
except ValueError as e:
    print(str(e))
```

```python
pad2_img = tf.image.pad_to_bounding_box(crop_img, offset_height=10, offset_width=10,
                                        target_height=70, target_width=70)
plt.imshow(pad2_img)
    
```
